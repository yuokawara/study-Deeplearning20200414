## 入力変数と出力変数の切り分け

```python
df.head()
```

```python
## df.iloc[行, 列]
df.iloc[0, 0]
```

```python
## 全ての行をとる場合 例：全ての行の０列目
df.iloc[:, 0]
```

```python
## 最後の列の1つ手前までデータをとる
X = df.iloc[:, :-1]
```

```python
y = df.iloc[:,  -1]
```

```python
y
```

## モデルの構築、検証

```python
from sklearn.linear_model import LinearRegression
```

```python
## モデルの宣言
model = LinearRegression()
```

```python
## モデルの学習
model.fit(X, y)
```

```python
## 検証　決定係数の計算
model.score(X, y)
```

## 訓練データ(train)と検証データ(test)

```python
from sklearn.model_selection import train_test_split
```

```python
## 訓練データと検証データの分割
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1) ## randam stateは乱数のシードを固定（再現性の確保）
```

```python
## モデルの学習
model.fit(X_train, y_train)
```

```python
##  検証　<- 検証データ
model.score(X_test, y_test)
```

```python
## 検証　<- 訓練データ
model.score(X_train, y_train)
```

## 予測値の計算とモデルの保存

```python
## 予測値
x = X.iloc[0,:]
```

```python
x
```

```python
y_pred = model.predict([x])[0]
y_pred
```

## モデルの保存

```python
from sklearn.externals import joblib
```

```python
## モデルの保存
joblib.dump(model, 'model_pkl')
```
